## Importing Related Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

## Loading The Data

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

## Data Cleaning

```python
age_median = train_df["Age"].median() 
age_median
```

```python
embarked_mode = train_df["Embarked"].mode()[0]
embarked_mode
```

```python
train_df['Age'] = train_df['Age'].fillna(age_median) # filling 'Age' column with it's median
train_df.drop('Cabin', axis=1, inplace=True) # dropping 'Cabin' column because it has too much null values
train_df['Embarked'] = train_df['Embarked'].fillna(embarked_mode)
```

```python
train_df.isnull().sum() # no more missing values now
```

## Feature Engineering

```python
train_df = train_df.drop(['PassengerId', 'Name', 'Ticket'], axis=1) # dropping columns that dont help with the prediction
train_df = pd.get_dummies(train_df, columns=['Sex', 'Embarked'], drop_first=True) # one-hot encoding
train_df
```

```python
X = train_df.drop('Survived', axis=1) # features
y = train_df['Survived'] # target
```

## Train-Test Split

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=31)
```

## Train the Logistic Regression Model

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

## Evaluate the Model

```python
y_pred = model.predict(X_val)
print("Accuracy:", accuracy_score(y_val, y_pred))
```

```python
print("Classification Report:", classification_report(y_val, y_pred))
```

```python
print("Confusion Matrix:", confusion_matrix(y_val, y_pred))
```
